````python
import json

# Minify mutation_map.json
with open('data/mutation_map.json') as f:
    data = json.load(f)

with open('data/mutation_map.min.json', 'w') as f:
    json.dump(data, f, separators=(',', ':'))

print("Done! Check your data/ folder for mutation_map.min.json")
```
````
